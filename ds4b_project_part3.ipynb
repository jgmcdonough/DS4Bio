###1.  Introduction
This dataset is a meta-analysis, with the goal of investigating whether or not restored oyster reefs have enhanced recruitment of nekton groups compared to controlled, unrestored reefs. Nekton groups are marine oragnisms that can swim freely [source](https://www.sciencedirect.com/topics/earth-and-planetary-sciences/nekton). This is measured by LRRs, or log response ratios. This is an important study to understand if restored reefs are a good management strategy and actually can be restored (and therefore worth the investment to restore reefs).

* [data processing](https://https://www.bco-dmo.org/dataset/845755): "An LRR with CIs > 0 implies that nekton recruitment was enhanced by oyster reefs, an LRR with CIs < 0 implies the opposite, and an LRR with CIs that include 0 implies no difference in recruitment between oyster reef and control habitats"

## Section 1: Data and research questions
#### Describe the data:
**Observations:**
The observations are nekton fish groups  observed at a particular oyster reef from different research publications. There are 1,848 observations of those observed nekton groups, from 25 different research publications. This specific paper is a meta-analysis, where the researchers collected data from other papers, checking for specified criteria to be able to compare them all (such as did authors quantify nekton abundance, unconstructed mud/sand habitats, and criteria relating to the makeup of the oyster reef itself).


**Features:**
The features are the information provided about the observation, such as the author/paper the data is from, the nekton group reported, the classification of the nekton group (resident or transient), the scientific name of the nekton group, the reef size, reef age, tidal elevation, abundance of nekton species, the latitude/longitude, the state, year, ocean, etc. There are 31 features for each observation. These features allow the researchers to understand the makeup and distribution of what fish/nekton groups are at each oyster reef system.

**Other data or knowledge are used?**
Other data/knowledge that is used is the ecological significance of the nekton groups the study decided to focus on. They could’ve just recorded every single observed organism that visited the reef, but chose to focus on specific species, ones that presumably have significant ecological impact, and their presence indicates a successful habitat restoration.


**Describe the research question the authors were asking: what specific hypothesis did they have and  and why was this data appropriate to investigate that hypothesis**
The researchers wanted to do a meta-analysis to look at answering questions about habitat restoration - specifically, they wanted to analyze the influence of the reef features/characteristics on the recruitment of nekton. In other words, does reef restoration actually help ecosystem functioning? The data they collected is appropriate to investigate this because the data gives us a picture of what a particular restored reef looks like (reef age, tidal elevation, reef size and height, etc.), and what species/taxon groups it attracts to it (taxon, family, resident/transient group, density of fish). Specifically, the researchers used ‘LRR’, or log response ratio to “compare nekton recruitment to oyster reefs vs. unstructured control habitat” - in other words, is recruitment of specific nekton groups enhanced in a restored reef habitat? This is important for assessing the success of restoration efforts and management.

```python
# want to see the geographic range of the taxon observed
# this is appropriate for this feature and data set because it would be interesting to
# know if the nekton groups seen at the sites are rare for that geographic range
# aka the reef is recruiting new groups for the area.
# seeing a boxplot would allow us to look at the variation of the latitudinal range of the
# taxon, as well as the mean/where does the bulk of the data lie for each group?

sns.catplot(data=data, x='taxon', y = 'lat_dd', kind='box') # categorical boxplot

# even though drum was the most frequent taxon group observed, that box is smaller,
# indicating that taxon group has a smaller latitudinal range. We can see that the
# grunt taxon group has a really small observed range, while goby and snapper
# have a larger latitudinal range. this relates to fig 2 - for grunt, there weren't
# any grunts found at natural reefs, only restored, which might explain why the range/
# variation is so small.
```

```python
# looking at the SI data - cleaning to not include 'nd', only S and I for values
data = data[data['SI'] != 'nd'] # S=subtidal reef, I=intertidal reef

# i want to somewhat recreat one of their figures, and look at LRR for each taxon group
# and see the differences between intertidal and subtidal reefs. This is appropriate
# because it visualizes the main question of the data - we can see if there is
# nekton group enhancement for restored vs. natural reefs and if there is different enhancement
# for subtidal vs intertidal reefs. This would give insight for restoration efforts
# on the best location of a restored reef to enhance recruitment.
# creating plot to examine the relation of LRR to taxon for each tidal group

sns.catplot(data=data, x='taxon',y='LRR', kind='box', col = 'SI')
# catplot box plot with subtidal reefs on the left and intertidal reefs on the right

# we can see that the goby taxon group has a lot more enhancement to restored reefs
# in subtidal zones compared to intertidal zones - same with snappers. It seems that
# for intertidal zones, most of the LRRs fall around a mean of 0, indicating that for this zone
# nekton recruitment is the same between restored and natural reefs. This gives insight
# to the hypothesis of the paper because we're really only see LRRs >0 for subtidal reefs,
# so for restoration/management strategies, it might be good to target subtidal zones for
# restoration.
```
